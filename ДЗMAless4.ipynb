Тема “Предел функции”

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

1.	Предложить пример функции, не имеющей предела в нуле и в бесконечностях.

```python
def f(x):
    if np.fabs(x) < 1:
        return np.sin(1/x)
    else:
        return np.sin(x)
    
x = np.linspace(-10, 10, 200)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

2.	Привести пример функции, не имеющей предела в точке, но определенной в ней.

```python
def f(x):
    if x < 0:
        return 0
    else:
        return 1
    
x = np.linspace(-100, 100, 1000)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

3. Исследовать функцию  по плану:
Область задания и область значений
Нули функции и их кратность
Отрезки знакопостоянства
Интервалы монотонности
Четность функции
Ограниченность
Периодичность

```python
# постороим функцию

def f(x):
    return np.power(x, 3) - np.power(x, 2)

x = np.linspace(-100, 100, 1000)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

```python
x = np.linspace(-0.0001, 1.0001, 100)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

Область задания и область значений: Все вещественные числа

Нули функции и их кратность: 0 (кратность 2) и 1

Отрезки знакопостоянства: от - бесконечности до 0, от 0 до 1, от 1 до + бесконечности

Интервалы монотонности: от - бесконечности до 0, от 0 до 2/3, от 2/3 до + бесконечности

Четность функции: ни чётная, ни нечётная

Ограниченность: неограничена сверху и неограничена снизу

Периодичность: не периодична
