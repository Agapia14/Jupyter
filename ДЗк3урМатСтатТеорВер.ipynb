Даны значения зарплат из выборки выпускников:
100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 230, 24, 57, 55, 70, 75, 65, 84, 90, 150
Посчитайте (несмещённое) среднее квадратичное отклонение для данной выборки.
В поле ответа впишите одно число. Нецелые числа следует округлить до 8 знаков после запятой

```python
import numpy as np

salary = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30,230, 24, 57, 55, 70, 75, 65, 84, 90, 150]

# среднее арифметическое
mean_salary = sum(salary)/len(salary)
print(f'Среднее арифметическое: {mean_salary}')

# среднее квадратичное отклонение

sum_mean = 0
for element in salary:
    sum_mean = (element - mean_salary)**2
    sum_mean += sum_mean 
    

std_salary = np.sqrt(sum_mean/len(salary))
print(f'Среднее квадратичное отклонение {std_salary}')

# смещенная дисперсия
variance_salary = sum_mean/len(salary)
print(f'Смещенная дисперсия {variance_salary}')

# несмещенная дисперсия
variance_salary2 = sum_mean/(len(salary) - 1)
print(f'Несмещенная дисперсия {variance_salary2}')
```
